```python
# creating targetvariable and features variable
features=df.drop(["Species"],axis=1)

target=df["Species"]   
```

```python
X=features

y=target
```

```python
# spliting dataset into  train and test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X,y , test_size = 0.20, random_state =17,stratify = y)

print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.metrics import accuracy_score

def print_metrics(actual, prediced):
    print('accuracy_score is {}'. format(accuracy_score(actual, predicted)))
```

```python
from  sklearn.linear_model import LogisticRegression

from sklearn.model_selection import GridSearchCV
```

```python
lr = LogisticRegression()

# Define the parameter grid with correct capitalization for 'C'
param_grid = {'penalty': ['l1', 'l2'], 'C': [100, 150, 200]}

log_model = GridSearchCV(lr, param_grid, scoring='accuracy', cv=5)
log_model.fit(X_train, y_train)
```

```python
# to predict the classes
# Get the predicted classes
y_train_class_preds = log_model.predict(X_train)
y_test_class_preds = log_model.predict(X_test)
```

```python
y_train_class_preds
```

```python
# Evalution of training model

from sklearn.metrics import accuracy_score

accuracy_score(y_train,y_train_class_preds)  
```

```python
# Evalution of training model
accuracy_score(y_test,y_test_class_preds)  
```
